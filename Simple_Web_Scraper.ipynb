now we will write a scraper that will handle errors and This scraper you will give it what **tag** you want to retrive and it will return it or return None Object 

````python
from urllib.request import urlopen
from urllib.error import HTTPError,URLError
from bs4 import BeautifulSoup

def getTag(tag,url):
    try:
    
        html = urlopen(url)
    except HTTPError as e:
        print(f'The page is not found is {e}')
        return None
    except URLError as e:
        print('the server is not found')
        return None
    
    try:
        bs = BeautifulSoup(html.read(),'html.parser')
        tag_returned = bs.find(tag) # dynamically finds the tag stored in the variable
    except AttributeError as e:
        return None
    
    return tag_returned

tag = input("""give the tag you want to retirve
            for example: 
                body.h1""")
page = getTag(tag,'http://www.pythonscraping.com/pages/page1.html')

if page is None:
    print(f'{tag} could not be found')
else:
    print(page)
````

Now we build a web scrapers that can separate 2 tags based on their class so this web scraper use **BeautifulSoup** to grab all the **red text** but none of the **green text** `<span class="red">....</span>`

````python
from urllib.request import urlopen
from bs4 import BeautifulSoup

html = urlopen('http://www.pythonscraping.com/pages/warandpeace.html')
bs = BeautifulSoup(html.read(), 'html.parser')

red_text_list = bs.find_all('span',{'class' : 'red'})
for text in red_text_list:
    print(text)
````

This web scraper script extract the table inside this wikipedia page `https://en.wikipedia.org/wiki/Python_(programming_language)` the name of the table is **Summary of Python 3's built-in types**

it only take the table and ignore the rest

````python
from urllib.request import urlopen, Request
from urllib.error import HTTPError, URLError
from bs4 import BeautifulSoup

URL = 'https://en.wikipedia.org/wiki/Python_(programming_language)'

# Mimic a real browser request
req = Request(URL, headers={'User-Agent': 'Mozilla/5.0'})

try:
    html = urlopen(req)
except (HTTPError, URLError) as e:
    print(f'Failed to fetch page: {e}')
    exit()

bs = BeautifulSoup(html.read(), 'html.parser')
wikitables = bs.find_all('table', {'class': 'wikitable'})

if not wikitables:
    print('No wikitables found.')
else:
    for i, table in enumerate(wikitables, start=1):
        print(table.get_text())
````

More about `find_all` and `find()`

````python
from urllib.request import urlopen
from bs4 import BeautifulSoup

html = urlopen('http://www.pythonscraping.com/pages/warandpeace.html')
bs = BeautifulSoup(html.read(), 'html.parser')


headers_list = bs.find_all(['h1','h2','h3','h4','h5','h6']) # returns a list of all the header tags in a document

print(headers_list)

print(60*'='+'\n')

green_and_read_list = bs.find_all('span', {'class': ['green', 'red']}) # return both the green and red span tags in the HTML
for text in green_and_read_list:
    print(text.get_text())

print(60*'='+'\n')

nameList = bs.find_all(text='the prince')
print(nameList)

print(60*'='+'\n')

title = bs.find_all(id='title', class_='text')
print(title)

````

learn Navigating Trees `.children` vs `.descendants`

---

Applying the two methods to the <head> tag:

    head_tag.children: This will yield one item: the <title> tag itself.
        Output: <title>The Dormouse's story</title>
    head_tag.descendants: This will yield two items: the <title> tag and the text string "The Dormouse's story" which is the child of the <title> tag.
        Output: <title>The Dormouse's story</title>, "The Dormouse's story" 

Both methods return a generator object, which can be iterated over or converted to a list using list().

````python
from urllib.request import urlopen
from bs4 import BeautifulSoup

html = urlopen('http://www.pythonscraping.com/pages/page3.html')
bs = BeautifulSoup(html, 'html.parser')

for child in bs.find('table',{'id':'giftList'}).descendants: # you can use .descendants
    print(child)
````

````python
from urllib.request import urlopen
from bs4 import BeautifulSoup

html = urlopen('http://www.pythonscraping.com/pages/page3.html')
bs = BeautifulSoup(html, 'html.parser')

for child in bs.find('table',{'id':'giftList'}).children: # you can use .descendants
    print(child)
````

**Dealing with siblings**

So it finds the **first `<td>`** in the entire table — which is inside the **first data row** (`gift1`). So `.next_siblings` only gets the **remaining `<td>` tags in that same row**.

it shows only the `<td>` inside (`gift1`) not the other `<td>` because they are in (different level)

````python
from urllib.request import urlopen
from bs4 import BeautifulSoup

html = urlopen('http://www.pythonscraping.com/pages/page3.html')

bs = BeautifulSoup(html, 'html.parser')
for sibling in bs.find('table', {'id':'giftList'}).td.next_siblings:
    print(sibling)
````

Use `.parent` and `.previous_sibling` So we  want to know the price next to this image `img1.jpg` so in this script we will give it the image and tell it to 


### Visually Traced

```
<tr id="gift1">
    <td> orange </td>
    <td> $15.00 </td>          ← 3. .previous_sibling lands HERE
    <td>                       ← 2. .parent lands HERE
        <img src="img1.jpg"/>  ← 1. find() starts HERE

```

So reading it **backwards**:

img → parent(**td**) → previous_sibling(**td**) → get_text() → "$15.00"

````python
from urllib.request import urlopen
from bs4 import BeautifulSoup
html = urlopen('http://www.pythonscraping.com/pages/page3.html')
bs = BeautifulSoup(html, 'html.parser')

print(bs.find('img',{'src':'../img/gifts/img1.jpg'}).parent.previous_sibling.get_text()) # or you can use `find_previous_sibling('td')`
````

Using **Regular Expressions** and **BeautifulSoup** 

So the page that we scraping right now has many product images and each **image** take the following form : `<img src="../img/gifts/img3.jpg">`

So if we want to grab **URLs** of all the **product images** in the **page** but we can't use `find_all("img")` because if we use it it will show nothing because modern websites have hidden images, blank images used for spacing and aligning elements,and other random image tags you might not be aware of.

and we don't depend on ***position*** of the image in the page in order to find the **correct tag**. because ***layout*** of the page **might change** 

So this is when we use ***Regular Expressions***  

````python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re

html = urlopen('http://www.pythonscraping.com/pages/page3.html')
bs = BeautifulSoup(html, 'html.parser')

images = bs.find_all('img',{'src':re.compile(r'\.\./img/gifts/img.*\.jpg')})
for image in images:
    print(image['src'])
````

Or you can do this

````python
from urllib.request import urlopen, Request
from urllib.error import HTTPError, URLError
from bs4 import BeautifulSoup

#URL = 'https://en.wikipedia.org/wiki/Python_(programming_language)'

URL = 'https://www.pythonscraping.com/pages/page3.html'

# Mimic a real browser request
req = Request(URL, headers={'User-Agent': 'Mozilla/5.0'})

try:
    html = urlopen(req)
except (HTTPError, URLError) as e:
    print(f'Failed to fetch page: {e}')
    exit()

bs = BeautifulSoup(html.read(), 'html.parser')

for link in bs.find_all('img'):
    if 'src' in link.attrs:
        print(link.attrs['src'])

````

**lambda** 

So the common use of **lambda** is to be argument passed in to other functions.
for example **BeautifulSoup** allows you to pass certain types of functions as parameters into
the `find_all` function. but these functions must take **tag** object as an **argument** and return a **boolean** 

So that line will  store All **Tags** in the **HTML page** That Have Exactly **2 Attributes** in a list and put it in **tags** variable 
`<tr>` is a tag and `class` & `id` are **attributes**

````python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re

html = urlopen('http://www.pythonscraping.com/pages/page3.html')
bs = BeautifulSoup(html, 'html.parser')

tags = bs.find_all(lambda tag: len(tag.attrs) == 2)
for tag in tags:
    print(tag.attrs)
````
